```python
import fontstyle as font
import emojis as emoji
```

```python
def welcome_prompt():
    print(font.apply( "\t\t\t\t   ~~~ WELCOME TO PIZZA POINT ~~~~\n","bold/blue"))
    print(font.apply("\t\t\t\t\tMake Your Own Pizza!!","bold/blue"))
    print(font.apply("\t\t\t\tWe Are Here To Add Happiness In Your Life","bold/blue"))
    print(emoji.encode(":pizza:"*49))
    print(font.apply("""
    \nYou can get up to 6 toppings on a pizza, first topping  is free and additional topppings are $1.5 each.\n""","bold"))
```

```python
# price list
available_pizza = ["small", "medium", "large","xl"]
pizza_price = {"small": 10, "medium":15, "large":20,"xl":25}

size = []
pizzas_list=[]
pizza_cost = []
total_cost=[]

def menu(user_choice):
    global pizza_cost
    global size
    print(font.apply('*'*100,"bold"))
    print(font.apply("Pizza Menu:\n","bold/blue"))
    for i, item in enumerate(available_pizza):
        print(i + 1, "   The price of", item, " cheese pizza is :$", pizza_price[item])
    print("-" * 50)
    while True:
        if user_choice == "add" or user_choice == "a":
            size = input("\nPlease select pizza size.  small, medium, large,  xl:\t ").lower()
            if  size == "small":
                pizza_cost.append(pizza_price[size])
                size = "small"
                break
            elif  size == "medium":
                pizza_cost.append(pizza_price[size])
                size = "medium"
                break
            elif size == "large":
                pizza_cost.append(pizza_price[size])
                size = "large"
                break
            elif size == "xl":
                pizza_cost.append(pizza_price[size])
                size = "extra Large"
                break

            else:
                print(font.apply("Please enter valid input.","bold"))
                continue

        else:
            print(font.apply("You have selected invalid input,please try again.","bold"))
            continue
    return  user_choice
```

```python
available_toppings = ['pineapple', 'pepperoni','mushrooms','onions','artichokes','tomatoes','bacon','beef','ham','anchovies','chicken']
toppings_price ={'bacon':1.5,'beef':1.5,'ham':1.5,'anchovies':1.5,'chicken':1.5,'pineapple':1.5, 'pepperoni':1.5,'mushrooms':1.5,'onions':1.5,'artichokes':1.5,'tomatoes':1.5}


def toppings():
    global pizza_cost
    global size

    print(font.apply('*'*100,"bold"))
    print(font.apply("\nAvailable toppings :\n","bold/blue"))
    for x,item in enumerate(available_toppings):
        print(x+1,' ' ,item)
    while True:
        try:
            limit = int(input("\nHow many toppings do you want?  1,2,3,4,5,6: "))
            if limit < 0 or limit > 6 or limit == str:
                print(font.apply("Select up to 6 toppings per pizza.","bold"))
                continue
            else:
                print(f"\nYou have added {limit} toppings in you list.")
                break
        except ValueError:
            print(font.apply("Please enter a valid number.","bold"))
    while True:
        if limit == 0:
            break
        user_toppings = input("\nPlease enter you topping : ")

        if len(topping) == limit:
            break
        elif user_toppings in available_toppings:
            topping.append(user_toppings)
            pizza_cost.append(toppings_price[user_toppings])
            if len(topping) == limit:
                break
        else:
            print(font.apply("\nPlease select available topping.","bold"))
            continue

    pizza = (size, topping)
    pizzas_list.append(pizza)

    if limit >= 1 and limit <= 6:
        pizza_cost.remove(1.5)
        print("\nThe cost of pizza is: $",sum(pizza_cost))
#customer ordered pizza will be saved in text file
    write_text()
```

```python
def edit_pizza():
    read_text()
    while True:
        try:
            delete = []
            remove_list = int(input("\nEnter a pizza number to remove  : "))
            for i in pizzas_list:
                remove_index = remove_list - 1
                delete = pizzas_list.pop(remove_index)
                if len(delete[1])>= 1:
                    delete[1].pop(0)
                    for i in delete[1]:
                        pizza_cost.remove(toppings_price[i])
                    pizza_cost.remove(pizza_price[delete[0]])
                    write_text()
                    break
                else:
                    pizza_cost.remove(pizza_price[delete[0]])
                    write_text()
                    break
            if remove_list == "no" or "n":
                break
            else:
                continue
        except ValueError:
            print(font.apply("Please enter valid number.","bold"))
```

```python
def edit_toppings():
    while True:
        edit = input("\nWould you like to add / remove the toppings? add or remove or no:  ").lower()
        if edit == "remove":
            read_text()            
            try:
                user= int(input("Select the number of pizza."))
                user_index = user - 1
                item = pizzas_list[user_index][1]
                print(f"\nYour selected topping:  {(','.join(item))}")
                remove_topping = input("\nWhich topping would you like to remove?  ").lower()
                if remove_topping in item:
                    item.remove(remove_topping)
                    pizza_cost.remove(toppings_price[remove_topping])
                    print(f"\nYour updated pizza price is $ {sum(pizza_cost)}")
                    write_text()

                else:
                    print("\nPlease remove valid topping from selected topping.")
                    continue
            except ValueError:
                print(font.apply("Please enter valid pizza number.","bold"))

        elif edit == "add":
            read_text()
            try:
                user = int(input("Select the number of pizza."))
                user_index = user - 1
                item = pizzas_list[user_index][1]
                print(f"\nYour selected toppings:  {','.join(item)}")
                add_topping=input("\nWhich topping would you like to add?  ").lower()
                if add_topping in available_toppings:
                    item.append(add_topping)
                    pizza_cost.append(toppings_price[add_topping])
                    print(f"\nYour updated pizza price is $ {sum(pizza_cost)}")
                    write_text()

                else:
                    print(font.apply("\nPlease select valid topping from available toppings.","bold"))
                    continue
            except ValueError:
                print(font.apply("Please enter valid pizza number.","bold"))

        elif edit == "no":
            break
        else:
            print(font.apply("\nPlease enter valid input.","bold"))
            continue
```

```python
def write_text():
    with open('pizza.txt', "w") as f:
        count = 0
        for i, item in pizzas_list:
            count += 1
            if len(item) == 0:
                f.write(f" {count}.  {i} cheese pizza.\n")
            elif len(item) == 1:
                f.write(f" {count}.  {i}: {(','.join(item))} pizza.\n")
            else:
                f.write(f" {count}.  {i}: {(','.join(item))} pizza.\n")
```

```python
def read_text():
    with open('pizza.txt', 'r') as f:
        pizza =f.readlines()
        print(pizza)
```

```python
def main():
    welcome_prompt()
    global topping
    global total_cost
    print("*"*100)
    while True:
        topping=[]
        user_choice = input("\n Please select a/add to add pizza, q/quit to quit pizza  :\t").lower()
        if user_choice == "q" or user_choice == "quit":
            break
        elif user_choice == "a" or user_choice =="add":
            menu(user_choice)
        else:
            print(font.apply("please enter valid input.","bold"))
            continue
        toppings()

    print("*"*100)
    print(font.apply("\nPizza in basket:\n", "bold/blue"))
    
    count = 0
    for i, item in pizzas_list:
        count += 1
        if len(item) == 0:
            print(f" {count}.  {i} cheese pizza.\n")
        elif len(item) == 1:
            print(f" {count}.  {i}: {(','.join(item))} pizza.\n")
        else:
            print(f" {count}.  {i}: {(','.join(item))} pizza.\n")

    while True:
        ask_input = input("\n Edit the pizza and pizza toppings ?  yes/y or no/n : ").lower()
        if ask_input == "yes" or ask_input == "y":
            ask_user =input("Edit the PIZZA / TOPPINGS ?  pizza or toppings  : ").lower()
            if ask_user == "pizza" :
                print(font.apply("\nYour pizza in basket: ","bold/underline"))
                edit_pizza()
                edit_toppings()
                write_text()
            elif ask_user == "toppings":
                edit_toppings()
                write_text()
        elif ask_input == "no" or ask_input == "n" :
            break
        else:
            print(font.apply("Enter valid answer.","bold"))
            continue

    print(font.apply("\n YOUR FINAL ORDER:\n","bold/blue"))
    count = 0
    for  i, item in  pizzas_list:
        count += 1
        if len(item) == 0:
            print(f" {count}.  {i} cheese pizza.\n")
        elif len(item) == 1:
            print(f" {count}.  {i}: {(','.join(item))} pizza.\n")
        else:
            print(f" {count}.  {i}: {(','.join(item))} pizza.\n")

    total_cost = sum(pizza_cost)
    print("Total cost of order :$", total_cost,"\n")
    print(emoji.encode(font.apply("Thanks for visiting!!! :smile:","bold/blue")))

if __name__ =="__main__":
    main()
```
